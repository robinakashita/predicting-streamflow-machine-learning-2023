```python
#Separate the Null Values
test_data = df[df["Tautuapae"].isnull()]
```

```python
df.dropna(inplace=True)
```

```python
#Create "X_train & Y_train from dataframe (df)

#Create the y_train /does not have Null values
# 'y_train' means Rows from the df['Flowrate'] with Non NULL values
y_train = df["Tautuapae"]
```

```python
y_train
```

```python
y_train.shape
```

```python
#X_train means Dataset except df['Tautuapae'] features with Non NULL Values
X_train = df.drop("Tautuapae", axis=1)
```

```python
X_train.shape
```

```python
#Build the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#Train the model on train Dataset (X_train, Y_train)
lr.fit(X_train,y_train)
```

```python
#Create the X test from the test Data
X_test = test_data.drop("Tautuapae", axis=1)
```

```python
X_test
```

```python
X_test.shape
```

```python
#Apply the trained model on 'X_test'
y_pred = lr.predict(X_test)
```
